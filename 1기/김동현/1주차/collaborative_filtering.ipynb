```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
#Data Creation

columns = ['Game A', 'Game B', 'Game C', 'Game D', 'Game E', 'Game F', 'Game G']
np.random.seed(42)

df = pd.DataFrame(np.random.choice([np.nan, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0], size=(1000, len(columns))), columns=columns)

display(df)
```

```python
from scipy.spatial.distance import cosine

#4번 사용자의 Game A 평점 계산

user_ratings = df.iloc[4]

similarities = df.apply(lambda row: 1 - cosine(row, user_ratings), axis=1)

top_users = similarities.nlargest(6).index
top_users = top_users[top_users != 4]  # 자기 자신은 제외

predicted_ratings = df.iloc[top_users].mean()

predicted_ratings['Game A']
```
